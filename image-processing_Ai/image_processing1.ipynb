```python
img= cv2.imread("./1.jpg")

#show image
cv2.imshow("RGB _image",img)
cv2.waitKey(0)

#resize             #طول  عرض
img = cv2.resize(img,(500,600))
#با ری سایز میتونیم هم بزرگ کنیم هم کوچیک
#show agein
cv2.imshow("RGB_image",img)
cv2.waitKey(0)

#متن دل به خواه و افزودن به تصویر
#put_text
font=cv2.FONT_HERSHEY_SIMPLEX                   #فونت اماده
bottomLeftCornerOfText=(30,30)          #جای متن
fontscale=0.5         #سایز فونت
fontcolor=(0,0,0)     #رنگ خط
linetype=2                #نوع خط

cv2.putText(img,"Arian Ghaderi",
            bottomLeftCornerOfText,
            fontscale,
            fontcolor,
            linetype)


#RGB --> GrayScale          #عکس رنگی به سیاه سفید
gray= cv2.cvtColor(img,cv2.COLOR_BGR2BGRA)

#show image
cv2.imshow("RGB_image",img)
cv2.imshow("GRAY_image",gray)
cv2.waitKey(0)
```
